#### Days employed

* Пропуски содержаться явно в одних и тех же строках.
* Можно будет заменить пропуски медианными значениями по группам.
* Перед этим ещё раз посмотрим на медианные значения по каждой группе.

```python
# Медианные значения в столбце "days_employed" для каждой группы `income_type`
df.groupby('income_type')['days_employed'].median()
```

* Для безработных и пенсионеров количество дней трудоустройва явно записано в часах, а не в днях
* У остальных категорий данные записаны как отрицательные значения. 

```python
# Диагностика ошибки значений в столбце "days_employed"
# Если сумма столбца будет равно 0, то это будет указывать на наличие центрирования значений по средней.
df['days_employed'].sum()
```

* Проверена гипотеза центрирования среднего значения, сумма столбца не равно 0, значит эта ошибка человеческого характера.
* Далее переведем значения в колонке "days_employed" в дни.

```python
# Создадим функцию перевода часов работы в дни в колонке "days_employed"
def days_employed(row):
    if row > 180000:
        return row/24
    else:    
        return row
```

```python
# Тестирование функции days_employed
print(days_employed(340266))
print(days_employed(0))
print(days_employed(-250))
```

```python
# Перевод часов в дни в колонке "days_employed"
df['days_employed'] = df['days_employed'].apply(days_employed)
```

```python
# Замена отрицательных значений в колонке "days_employed" положительными
df['days_employed'] = df['days_employed'].abs()
```

```python
# Проверим медианные значения 'days_employed' для каждой группы 'income_type'
df.groupby('income_type')['days_employed'].median()
```

* Теперь колонка хранит только положительные значения в одном диапазоне и можно заменять пропуски медианными значениями по каждой категории для точечного изменения дисперсии.

```python
# Заменим пропуски в 'days_employed' медианными значениями по группе  'income_type'
df['days_employed_median'] = df.groupby('income_type')['days_employed'].transform('median')
df['days_employed'].fillna(df['days_employed_median'], inplace=True)
df.drop(['days_employed_median'], axis=1, inplace=True)
```

```python
# Посмотрим на изменения медианных значений в колонке days_employed
df.groupby('income_type')['days_employed'].median()
```

### Категоризация целей кредита

#### Лемматизация 

* В связи с тем, что цели кредита повторялись между клиентами, но имели разные формулировки - необходимо лемматизировать колонку, чтобы объединить цели в категории.

```python
# Создадим колонку 'purpose_lem', куда поместим лемматизированные значения
df['purpose_lem'] = df['purpose']
```

```python
# Создадим функцию по лемматизации столбца
m = Mystem() 
def lem(text):
    return ' '.join(m.lemmatize(text))
```

```python
text = 'на свадьбу'
lem(text)
```

```python
# Создадим колонку с лемматизированными значениями
df['purpose_lem'] = df['purpose'].apply(lem)
```

```python
#Новая таблица df_lem без столбца 'purpose'
df_lem = df.drop(columns=['purpose'], axis=1)
```

```python
# Посмоторим на новую таблицу
df_lem = df_lem.drop(columns=['Unnamed: 0'], axis=1)
```

```python
# Посмоторим на новую таблицу
df_lem.head()
```

## Влияние факторов на возраст кредита в срок

### Наличие детей

* Создадим функцию, которая будет создавать сводную таблицу для каждого фактора и возращать подсчёт процента возрата заплативших в срок и должников.

```python
def pivot_table(column):
    table = df_lem.pivot_table(index=[column], columns='debt', values='education', aggfunc='count')
    table = table.fillna(0)
    table['total'] = table[0] + table[1]
    table['paid'] = table[0]/table['total']*100
    table['debt'] = table[1]/table['total']*100
    table_final = table[['paid', 'debt']]
    return table_final
    
```

```python
# Построим график
pivot_table('children')\
.plot(kind='bar', figsize=(10,7))

plt.ylabel('%')
plt.xlabel('')
plt.title('Зависимость между платёжеспособностью клиентов и количеством детей', fontsize=16)
plt.show()

pivot_table('children')
```

**Вывод**

* Вне зависимости от количества детей в семье возрат по кредиту во всех семьях колеблеться в пределах 90,24% - 100%. * Максимальный возраст по кредиту был показан для семей с пятью детьми, среди которых не было должников.
