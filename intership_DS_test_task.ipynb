```python
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Функция AB_1 подсчитывает число покупателей, попадающих в каждую группу,
# Нумерация ID сквозная и начинается с 0. На вход функция получает целое число n_customers (количество клиентов).


def AB_1(num):
  l = np.linspace(0, num, num+1, dtype='int') # Зададим список ID клиентов, начиная с нуля со сковзной нумерацией


  def summa(n): # Функция summa возращает сумму значений целого числа
    b = []
    while n>0:
      b.append(n%10)
      n = n//10   
    return np.sum(b)
  
  return Counter([summa(i) for i in l]) # Получим количество попупателей в каждой из групп


```

```python
# Получим значения функции AB_1 для 1000 покупателей
print("Количество групп для 1000 покупатетелей:", len(AB_1(1000)))
```

```python
# Получим значения функции AB_1 для 1000 покупателей
plt.figure(figsize=(10,8))
df = pd.DataFrame(pd.Series(AB_1(1000)).reset_index())
plt.bar(df['index'], df[0], color = 'brown')

plt.title('Распределение количества покупателей по группам', fontsize=16)
plt.xlabel('Количество групп', fontsize=14)
plt.ylabel('Количество посетителей в каждой группе', fontsize=14)
plt.xticks(np.arange(0,30, step=2))
None
```

```python
# Функция AB_2 подсчитывает число покупателей, попадающих в каждую группу,
# Нумерация ID сквозная и начинается с произвольного числа. На вход функция получает целое число n_customers (количество клиентов) и произвольное число.


def AB_2(id, num):
  last = id + num
  border = last + 1
  l = np.linspace(id, last,border, dtype='int') # Зададим список ID клиентов, начиная с нуля со сковзной нумерацией


  def summa(n): # Функция summa возращает сумму значений целого числа
    b = []
    while n>0:
      b.append(n%10)
      n = n//10   
    return np.sum(b)
  
  return Counter([summa(i) for i in l]) # Получим количество попупателей в каждой из групп


```

```python
# Получим значения функции AB_2 для 1000 покупателей
print("Количество групп для 1000 покупатетелей:", len(AB_2(12345, 1000)))
```

```python
# Построим график распеделения количества покупателей, где ID задан случайно
plt.figure(figsize=(10,8))
df = pd.DataFrame(pd.Series(AB_2(12345, 1000)).reset_index())
plt.bar(df['index'], df[0], color = 'rosybrown')

plt.title('Распределение количества покупателей по группам', fontsize=16)
plt.xlabel('Количество групп', fontsize=14)
plt.xticks(np.arange(0,30, step=2))
plt.ylabel('Количество посетителей в каждой группе', fontsize=14)
None
```
