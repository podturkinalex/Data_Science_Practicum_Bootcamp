**Задача**
* Определить значимые параметры и отследить аномалии.

*Описание данных*

* Исторические данные объявлений о продажах квартир в Санкт-Петербурге и его алгомерации из архива  сервиса Яндекс.Недвижимость.


* Часть данных была предоставлена пользователями, а другая часть получена автоматически на основе картографических данных.

*Согласно документации к данным:*
*	airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
*	balcony — число балконов
*	ceiling_height — высота потолков (м)
*	cityCenters_nearest — расстояние до центра города (м)
*	days_exposition — сколько дней было размещено объявление (от публикации до снятия)
*	first_day_exposition — дата публикации
*	floor — этаж
*	floors_total — всего этажей в доме
*	is_apartment — апартаменты (булев тип)
*	kitchen_area — площадь кухни в квадратных метрах (м²)
*	last_price — цена на момент снятия с публикации
*	living_area — жилая площадь в квадратных метрах (м²)
*	locality_name — название населённого пункта
*	open_plan — свободная планировка (булев тип)
*	parks_around3000 — число парков в радиусе 3 км
*	parks_nearest — расстояние до ближайшего парка (м)
*	ponds_around3000 — число водоёмов в радиусе 3 км
*	ponds_nearest — расстояние до ближайшего водоёма (м)
*	rooms — число комнат
*	studio — квартира-студия (булев тип)
*	total_area — площадь квартиры в квадратных метрах (м²)
*	total_images — число фотографий квартиры в объявлении

* Посмотрим на распределение количества комнат в квартирах

* Основной дипазон находится для одна-7 комнатных квартир.
* Далее количество объявлений с комнатами более 8 комнат снижается в прогрессии.
* Посмотрим есть ли связь между количеством комнат и общей площадью квартиры.

```python
# Построим график зависимости количества комнат и площади
df.pivot_table(index='rooms', values='total_area', aggfunc='median')\
.plot(y='total_area', figsize=(10,7), label='Общая площадь', linewidth=4)

plt.xlabel('Количество комнат', fontsize=12)
plt.ylabel('Медианное значение общей площади, м²',  fontsize=12)
plt.title('Зависимость общей площади от количества комнат', fontsize=16)
plt.show()
```

* На графике видно, что начиная от 0 до 10 комнатных квартир их площадь равномерно растет.
* Начиная с 12 комнатных квартир общая площадь вырастает на несколько порядоков, что выглядит аномально.
* Проверка предложений в интернете не увенчалась успехов для квартир более 10 комнат. Вероятно это не настоящие предложения.
* Нулевые квартиры - это могут быть квартиры студии или квартиры со свободной планировкой. Далее буду проверять их метраж.

Везде, где есть студия площадь кухни не заполнена, значит площадь кухонь здесь можно заменить на нули.

```python
#Тестирование функции замены пропущенных значений в колонке кухня для квартир студий
def kitchen_area_zero(row):
    studio = row['studio']
    kitchen_area = row['kitchen_area']
    if studio == True:
        return 0
    else:
        return kitchen_area

row_values = [True, 25]
row_columns = ['studio', 'kitchen_area']
row = pd.Series(data=row_values, index=row_columns) 

kitchen_area_zero(row)    
```

```python
# Высота потолка для квартир менее 80 м2
# Медианное значение
print('Медианное значение высоты потолка для квартир с площадью менее 80 м2:')
print(df.query('total_area<80 and ceiling_height<10 and ceiling_height>2')['ceiling_height'].median())

# Гистограмма
plt.figure(figsize=(8,7))
sns.histplot(df.query('total_area<80 and ceiling_height<10 and ceiling_height>2')['ceiling_height'], bins=10)
plt.title('Распределение высоты потолка для квартир с площадью менее 80 м2')
plt.xlabel('Высота потолка, м')
plt.ylabel('Частота')
plt.show()

```

 * Здесь мы видим более скученные данные и более 10 тыс раз встречается значения около 2,6 метров
 * Предлагаю заменить пропуски только для этой выборки для квартир с площадью менее 80 кв м.

```python
# Тестирование функции замены пустых значений в колонке ceiling_height медианой для квартир менее 80 м2
def ceiling_median(row):
    ceil_median = df.query('total_area<80 and ceiling_height<10 and ceiling_height>2')['ceiling_height'].median()
    total_area = row['total_area']
    ceiling_height = row['ceiling_height']
    if total_area<80:
        return ceil_median
    else:
        return ceiling_height
    
row_values = [81, 1]
row_columns = ['total_area', 'ceiling_height']
row = pd.Series(data=row_values, index=row_columns) 

ceiling_median(row)
```

```python
#Применение функции ceiling_median
df.loc[df.ceiling_height.isna(), 'ceiling_height'] = df[df.ceiling_height.isna()].apply(ceiling_median, axis=1)
df['ceiling_height'].isna().sum()
```

* Осталось 1333 пропуска.
* Создадим отдельно колонку, которая будет хранить аномально высокие значения потолка выше 10 м и ниже 2 м.

```python
# Тестирование функции проверки аномальной высоты потолка.
# Если аномалия - больше 10 или менее 2х метров - то True, если меньше, то False.
def is_ceil_anormal(row):
    if row<2 or row>10:
        return True
    else:
        return False
print(is_ceil_anormal(11))
print(is_ceil_anormal(1.9))
print(is_ceil_anormal(2.5))
```

```python
df['is_ceiling_height_anormal'] = df['ceiling_height'].apply(is_ceil_anormal)
```
