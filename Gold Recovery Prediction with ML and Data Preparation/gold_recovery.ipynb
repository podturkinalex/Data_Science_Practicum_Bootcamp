# Восстановление золота из руды

* Данные предоставлены https://www.zyfra.com/ . `Zyfra` предоставляет решения для эффективной работы промышленных предприятий.

**Цель**
* Предсказать коэффициент восстановления золота из золотосодержащей руды.

**Задачи**
* Предобработать данные и проанализировать данные разных этапов технологических процессов восстановления золота из руды.
* Разработать прототип модели машинного обучения с кастомной метрикой оценки качества модели. В качестве метрики необходимо построить функцию подсчёта `Symmetric Mean Absolute Percentage Error` (сокращенно `sMAPE`) для расчёта эффективностей обогащения чернового `rougher.output.recovery` и финального продуктов `final.output.recovery`.
* Обучить модель с наименьшим значением `sMAPE` на тестовых данных.

**Описание данных**

* Данные храняться в трёх таблицах `gold_recovery_full_new.csv`, `gold_recovery_train_new.csv`, `gold_recovery_test_new.csv`.

* Два целевых отклика: `rougher.output.recovery`, `final.output.recovery`.

*Технологический процесс*
*	Rougher feed — исходное сырье
*	Rougher additions (или reagent additions) — флотационные реагенты:

1)	*Xanthate* — ксантогенат (промотер, или активатор флотации);

2)	*Sulfate* — сульфат (на данном производстве сульфид натрия);

3)	*Depressant* — депрессант (силикат натрия).

*	Rougher process (англ. «грубый процесс») — флотация
*	Rougher tails — отвальные хвосты
*	Float banks — флотационная установка
*	Cleaner process — очистка
*	Rougher Au — черновой концентрат золота
*	Final Au — финальный концентрат золота

*Параметры этапов*
*	air amount — объём воздуха _air
*	fluid levels — уровень жидкости _level
*	feed size — размер гранул сырья _size
*	feed rate — скорость подачи _rate

*Наименование признаков*

Наименование признаков должно быть такое:
`[этап].[тип_параметра].[название_параметра]`

Пример: `rougher.input.feed_ag`

*Возможные значения для блока* `[этап]`:
*	rougher — флотация
*	primary_cleaner — первичная очистка
*	secondary_cleaner — вторичная очистка
*	final — финальные характеристики

*Возможные значения для блока* `[тип_параметра]`:
*	input — параметры сырья
*	output — параметры продукта
*	state — параметры, характеризующие текущее состояние этапа
*	calculation — расчётные характеристики

### Проверка расчетов эффективности обогащения

```python
#Создадим функцию recovery_rougher для оценки эффективность обогащения после флотации
def recovery_rougher(row):
    C = row['rougher.output.concentrate_au']
    F = row['rougher.input.feed_au']
    T = row['rougher.output.tail_au']
    Recovery = ((C*(F-T))/(F*(C-T)))*100
    return Recovery
```

```python
#Применим функцию
train['rougher.output.recovery_2'] = train.apply(recovery_rougher, axis=1)
```

```python
# Проверим расчёты
col = ['rougher.output.recovery', 'rougher.output.recovery_2']
train[col].head()
```

```python
#Уберем из выборки лишний столбец rougher.output.recovery_2'
train = train.drop('rougher.output.recovery_2', axis=1)
```

* Расчёты были сделаны верно. Ошибка `MAE` нулевая.
* При расчётах `rougher.output.recovery` не учитывается концентрация веществ в других металлах, как Серебро. Значит можно расчитывать эти метрики по отдельности без суммирования концентраций с другими металлами.

#### Функция для вычисления итоговой Symmetric Mean Absolute Percentage Error

* Создадим функцию расчёта `Symmetric Mean Absolute Percentage Error` (сокращенно `sMAPE`) для расчёта эффективностей обогащения чернового `rougher.output.recovery` и финального продуктов `final.output.recovery`.

```python
def sMAPE(train, predict):
    sums = np.sum(np.abs(train - predict)/((np.abs(train)+np.abs(predict))/2))
    smape = 1/len(train)*sums*100
    return smape     
           
```

```python
#Проведем тестиварование функции
target_train_1 = pd.Series([25,1,1])
target_predict_1 = pd.Series([25, 1,1])

sMAPE(target_train_1, target_predict_1)
```

* Функция работает.
* Теперь создадим функцию sMPAPE_f для подсчёта Итогового sMPAPE.

```python
#Создадим функцию для подсчёта Итоговой sMAPE
def sMAPE_f(target_train, target_predict):
   
       
    sMAPE1 = sMAPE(target_train[:,0], target_predict[:,0])
    sMAPE2 = sMAPE(target_train[:,1], target_predict[:,1])
    smape_fin = 0.25*sMAPE1 + 0.75*sMAPE2
    return smape_fin
```
