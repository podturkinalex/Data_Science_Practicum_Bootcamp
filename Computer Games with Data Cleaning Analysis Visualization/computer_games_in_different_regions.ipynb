**Цель**

**Описание данных**

Исторические данные о продажах игр в четырёх регионах.
*	`Name` — название игры
*	`Platform` — платформа
*	`Year_of_Release` — год выпуска`*`

 `*` *В 2016 году данные могут быть неполные*.

**Пропуски в колонке `user_score`**

```python
# Посмотрим на уникальные значения колонки user_score
games['user_score'].unique()
```

* Встречается как пропуски, так и нечисловое значение `tbd`
* Согласно источнику metacritic `tbd` обозначается для рейтинга игр, по которым ещё нет достаточного количества отзывов.  https://www.metacritic.com/faq#:~:text=every%20summary%20page.-,What%20does%20%22tbd%22%20mean%3F,4)%20reviews%20in%20our%20database.


```python
#Посмотрим на количество строк со значением "tbd" в колонке user_score
len(games.query('user_score=="tbd"', engine='python'))
```

* Всего 6605 пропусков в колонке `user_score`. Дополнительно есть 2376 ячейки имеют значение "tbd"
* Посмотрим чему соответствует значение "tbd".

```python
#Посмотрим какие категории rating встречаются для значения 'user_score=="tbd"'
games.query('user_score=="tbd"', engine='python')['rating'].value_counts()
```

*Лидирует рейтинг "е", затем "t" и замыкает "ec".*

```python
#Посмотрим какие категории rating встречаются для значения 'user_score' в целом.
games['rating'].value_counts()
```

```python
#Посмотрим как категории рейтинга встречаются для пропущенных значений колонки user_score
games.query('user_score.isna()', engine='python')['rating'].value_counts()
```

*Снова на первом месте рейтинг 'e', потом 't', 'e10+' и 'm' поменялись местами и 'ес' гораздо меньше, чем остальных.*

* Данных по рейтингу игры для пропущенных значений в колонке `user_score` немного.
* Однако характер распределения рейтингов в целом схожий между общей выборкой, так и между "tbd" и пропусками. Скорее всего пропуски в колонке `user_score` указывают на то, что оценки пользователей ещё не были поставлены (не были определены).
* Заменить рейтинг медианными значениями не имеет смысла, т.к. пропусков слишком много и замена приведет к сильному искажению данных.
* Нужно заменить пустые значения и `tbd` в колонке `user_score` на пропуски, чтобы потом можно было все пропуски заменить на единое число и фильтровать данные.

```python
#Замена tbd в колонке user_score на пустые значения
games['user_score'] = games['user_score'].replace({"tbd":np.nan})
games['user_score'] = games['user_score'].fillna(-100)

# Заменим формат колонки 'user_score' на float
games = games.astype({'user_score':"float64"})
```

```python
games.info()
```

```python
# Построим сводную таблицу стандартного отклонения оценкой критиков для каждой игры
std_user_score = (games
 .query('user_score!=-100', engine='python')
 .pivot_table(index='name', values='user_score', aggfunc='std')
)
std_user_score.head()
```

```python
std_user_score['user_score'].describe()
```

```python
#Разброс стандртного отклонения оценки критиков для каждой игры
sns.set(rc={'figure.figsize':(7,6)})
sns.set_style("white")


sns.boxplot(y= std_user_score['user_score'], color='g')
plt.ylabel('Оценка пользователей, std')
plt.title('Разброс STD оценки пользователей для каждой игры')
plt.show()

#Описательная статистика для стандртного отклонения оценок пользователей для каждой игры
std_user_score['user_score'].describe()
```

*Вывод*
* Все пропуски и все значения `tbd` в колонке `user_score` были заменены на -100, чтобы перевести затем в формат `float`.
* Процент пропусков (-100) теперь увеличился и достиг 45% в колонке `user_score`.
* 15% данных также как и у критиков имеют большие разбросы между оценками пользователей по играм. Поэтому заменять пустые значения медианами по каждой игре может сильно исказить результат.
* Удаление такого большого процента пропусков сильно исказит данные. Т.к. пропуски нельзя заменить, а в соседних строках есть важная информация, то принято решение работать с оставшимися пропусками для `critic_score` и `user_score`.

**Пропуски к колонке `rating`**

```python
# Посмотрим на категории рейтингов
print('Количество уникальных категорий рейтингов:', games['rating'].nunique())
print()
print(games['rating'].value_counts())
```

```python
#Заменим пропуски на "нет категории"
games['rating']=games['rating'].fillna('Нет категории') 
games.isna().sum()
```

```python
def rating(row):
    if row=='E' or row=='K-A':
        return 'Все'
    if row=='T' or row=='E10+':
        return 'Подростки'
    if row=='M' or row=='AO':
        return 'Взрослые'
    if row=='EC':
        return 'Дети'
    if row=='RP':
        return 'Нет категории'
    else:
        return 'Нет категории'
rating('K-A')    
```

```python
games['rating_rus']=games['rating'].apply(rating)
games.head()
```

```python
# Посмотрим на распределения медианных оценок для 'xone', 'pc'
(games_2014
 .query('platform in @col and user_score>=0')
 .pivot_table(index='platform', values='user_score', aggfunc=['describe', 'var', 'count'])
)
```

```python
# Проверка распределений на нормальность при помощи теста Колмогорова-Смирнова для колонки platform pc и xone:
# Если распределение нормальное, то уровень значимости (2-е значение в строке) будет больше 0,05.
for name, groupname in games_2014.query('platform in @col and user_score>=0').groupby('platform'):
    print(name, stats.kstest(groupname['user_score'],'norm'))
```

* Медианное значение и максимум пользовательских оценок выше для PC, чем для Xbox One. 
* Размеры выборок и дисперсии различаются.
* Оба распределения очевидно ненормально распределены. Будем использовать mann whitney u test вместо t-test.


Гипотеза №1
* Сформулируем нулевую гипотезу: cредние пользовательские рейтинги платформ Xbox One и PC одинаковые.
* Т.к. медианные значения для PC выше, то сформулируем альтернативную гипотезу -  cредние пользовательские рейтинги платформ PC выше Xbox One.     


```python
# Проверим гипотезу, что одностороннюю гипотезу, что cредние пользовательские рейтинги платформ PC выше Xbox One.     
alpha = .05 # уровень значимости
results = stats.mannwhitneyu(x=games_2014.query('platform =="PC" and user_score>=0')['user_score'],
                   y=games_2014.query('platform =="XOne" and user_score>=0')['user_score'],
                  alternative='greater') # результаты непарметрического теста

PC_median = games_2014.query('platform =="PC" and user_score>=0')['user_score'].median()
XOne_median = games_2014.query('platform =="XOne" and user_score>=0')['user_score'].median()

print('Медианное значение пользовательских оценок PC', PC_median)
print('Медианное значение пользовательских оценок Xone', XOne_median)
print(f'p-значение {results.pvalue}')
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу, что cредние пользовательские рейтинги платформ Xbox One и PC одинаковые.")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Вывод
* Не отвергаем нулевую гипотезу. Медианные значения между пользовательскими оценками не различаются для PC и платформы XOne.

```python
# Проверка распределений на нормальность при помощи теста Колмогорова-Смирнова для колонки genre для 'action', 'sports':
# Если распределение нормальное, то уровень значимости (2-е значение в строке) будет больше 0,05.
for name, groupname in games_2014.query('genre in @col and user_score>=0').groupby('genre'):
    print(name, stats.kstest(groupname['user_score'],'norm'))
```

* Оба распределения очевидно ненормально распределены. Будем использовать mann whitney u test вместо t-test.

 Гипотеза №2

* Сформулируем нулевую гипотезу: средние пользовательские рейтинги жанров Action и Sports одинаковые.
* Т.к. медианные значения схожие, то мы не можем говорить о конкретных различиях между жанрами. Сформулируем альтернативную гипотезу - средние пользовательские рейтинги жанров Action и Sports различаются.  

```python
# Проверим гипотезу, что двухстороннюю гипотезу, что оценки между жанрами различаются.
alpha = .05 # уровень значимости
results = stats.mannwhitneyu(x=games_2014.query('genre =="Action" and user_score>=0')['user_score'],
                   y=games_2014.query('genre =="Sports" and user_score>=0')['user_score'],
                  alternative='two-sided') # результаты непарметрического теста

action_median = games_2014.query('genre =="Action" and user_score>=0')['user_score'].median()
sports_median = games_2014.query('genre =="Sports" and user_score>=0')['user_score'].median()

print('Медианное значение пользовательских оценок жанра Action', action_median)
print('Медианное значение пользовательских оценок жанра Sports', sports_median)
print(f'p-значение {results.pvalue}')
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу, что cредние пользовательские оценки жанров action и sports одинаковые.")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Вывод
* Отвергаем нулевую гипотезу, что cредние пользовательские оценки жанров Action и Sports одинаковые. Медианные значения пользовательских оценок для жанра Action выше, чем для жанра Sports и составляют 7,1 балла.
