**Цель**
* Выбрать наиболее перспективное местонахождение для добычи нефти нефтедобывающей компании.

**Задачи**

## Разделим и стандартизируем выборки

* Необходимо построить три модели для каждой выборки. Для удобства будет работать с данными по порядку их наименований.

#### Данные по Региону_0

```python
target_0 = region_0['product']

#Создадим таблицу наших переменных, по которым будет производится предсказание модели - все, кроме отлика и id
features_0 = region_0.drop(columns=['product', 'id'], axis=1)
```

```python
#Выделим 75% на тренировочную выборку и 25% на валидационную
def split(target, features):
    return train_test_split(features, target, test_size=0.25, random_state=0)           
```

```python
features_train_0, features_valid_0, target_train_0, target_valid_0 = split(target_0, features_0)
```

```python
print(features_train_0.shape)
print(features_valid_0.shape)
```

* Разделение произошло корректно.
* Проведем стандартизацию переменных.

```python
#Создадим функцию стандартизации, 
#которая вернет стандартизированные признаки для тренировочной и валидационной выборок
def standartscale(features_train, features_valid):
    #Список колонок, по которым будет проходить скалирование
    col = ['f0', 'f1', 'f2']
    pd.options.mode.chained_assignment = None
    scaler = StandardScaler()
    scaler.fit(features_train[col])
    features_train[col] = scaler.transform(features_train[col])
    features_valid[col] = scaler.transform(features_valid[col])
    return features_train, features_valid
```

```python
#Применим функцию standartscale
features_train_0, features_valid_0 = standartscale(features_train_0, features_valid_0)
features_train_0.head()
```

* Стандартизация прошла успешно.

#### Данные по Региону_1

```python
target_1 = region_1['product']

#Создадим таблицу наших переменных, по которым будет производится предсказание модели - все, кроме отлика и id
features_1 = region_1.drop(columns=['product', 'id'], axis=1)
```

### Обучение и проверка модели

#### Модель по Региону_0

```python
#Создадим функцию model, которая вернет массив с предсказаниями, среднее значение предсказаний и rmse
def model(features_train, target_train, features_valid, target_valid):
    model = LinearRegression()
    model.fit(features_train, target_train)
    predictions_valid = model.predict(features_valid)
    mean = predictions_valid.mean().round(2)
    rmse = (mean_squared_error(target_valid, predictions_valid))**0.5
    return predictions_valid, mean, rmse
```

```python
#Применение функции model для Региона_0
predictions_valid_0, mean_0, rmse_0 = model(features_train_0, target_train_0, features_valid_0, target_valid_0)
print('Средний запас предсказанного сырья в Регионе:', mean_0)
print('RMSE модели линейной регрессии на валидационной выборке в Регионе:', rmse_0)
```

* Сравним с пустой моделью

```python
#Создадим функцию для проверки для построения пустой средней модели
def mean_model(target_train, target_valid):
    predicted_valid_dummy = pd.Series(target_train.mean(), index=target_valid.index) #индексы по валидационной выборке
    mean_dummy = predicted_valid_dummy.mean()
    rmse_dummy = (mean_squared_error(target_valid, predicted_valid_dummy))**0.5
    return 'Средний запас предсказанного сырья в регионе:', mean_dummy, 'RMSE модели линейной регрессии на валидационной выборке в регионе:', rmse_dummy
```

```python
#Посчитаем средний запас и rmse для средней модели
mean_model(target_train_0, target_valid_0)
```

* RSME ниже на обученной модели почти на 6 тыс. баррелей, чем на средней модели, значит регрессионная модель предсказывает лучше.

```python
#Напишем функцию, которая по предсказаниям модели отберет 200 самых продуктивных точек в регионе
#и посчитает прибыль по данным из объема реальных скважин
def income(target_valid, predictions_valid, count):
    prob_sorted = predictions_valid.sort_values(ascending=False)
    selected = target_valid[prob_sorted.index][:count]
    revenue = REVENUE_FROM_1K_BARREL*(selected.sum())
    INCOME = revenue - BUDGET
    return INCOME
```

```python
#Расчитаем суммарную прибыль для 200 самых продуктивных скважин в Регионе_0 
INCOME_REGION_0 = income(target_valid_0, predictions_valid_0, 200)
round(INCOME_REGION_0, 4)
```

```python
#Расчитаем суммарную прибыль для 200 самых продуктивных скважин в Регионе_1 
INCOME_REGION_1 = income(target_valid_1, predictions_valid_1, 200)
round(INCOME_REGION_1, 4)
```

```python
#Расчитаем суммарную прибыль для 200 самых продуктивных скважин в Регионе_2
INCOME_REGION_2 = income(target_valid_2, predictions_valid_2, 200)
round(INCOME_REGION_2, 4)
```

## Расчёт прибыли и рисков 

### Применим технику Bootstrap, чтобы найти распределение прибыли.

#### Прибыль и риски в Регионе_0

```python
# Создадим функцию revenue_region, которая будет считать по предсказаниям модели прибыль по 200 лушчим точкам
# Будет использован метод Bootstrap для создания 1000 подвыборок для 200 точек из 500
# Функция на выходе выдаёт:
# среднюю прибыль в регионе, 95% доверительный интервал,  описательную статистику распределения и оценку рисков
def income_region(target_valid, predictions_valid):
    state = np.random.RandomState(0)
    incomes = []

    for i in range(1000):
        target_subsample = target_valid.sample(n=500, replace=True, random_state=state)
        probabilities_subsample = predictions_valid[target_subsample.index]
        incomes.append(income(target_subsample, probabilities_subsample, 200))
        
    incomes_region = pd.Series(incomes)
    
    #Гистограмма распределения прибыли в Регионе
    plt.figure(figsize=(10,6))
    #a = incomes_region.plot(kind='hist', title='Распределение прибыли в Регионе для 200 наиболее продуктивных скважин')
    a = sns.histplot(incomes_region)
    a.set_xlabel('Прибыль в регионе')
    a.set_ylabel('Частота')
    plt.title('Распределение прибыли в Регионе для 200 наиболее продуктивных скважин')
    plt.show()
    
    #Средняя по региону
    mean = round(incomes_region.mean(), 2)
    
    #Доверительный интервал
    low = incomes_region.quantile(0.0025).round(2)
    high = incomes_region.quantile(0.975).round(2)
        
    #Анализ рисков 
    risk = (((incomes_region < 0).mean())*100).round(2)
    
    return  low, high, mean, risk, incomes_region
   
```

```python
#Применим функцию income_region для расчёта прибыли и рисков в Регионе_0
low_0, high_0, mean_0,  risk_0, incomes_region_0 = income_region(target_valid_0, predictions_valid_0)
print('Суммарная прибыль с 200 точек в Регионе_0:', mean_0)
print()
print('Процент рисков отрицательной прибыли', round(risk_0, 3), "%")
print()
print(f'95%-ый доверительный интервал находится в интервале от {low_0} до {high_0} рублей')
```

* В `Регионе_0` **суммарная прибыль** у 200 лучших точек, отобранных моделью из 500 случайных точек, составляет `470832786.907217 рублей`, что выше заложенного бюджета.
* Суммарная прибыль с 95% уверенностью лежит в интервале от -319042557.2321199 до 997335233.4561398 рублей.
* **Риски убытоков** составляют `3,5 %`.
