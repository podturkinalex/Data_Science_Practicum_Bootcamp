## Предобработка данных

* Сначала создадим сводную таблицу `calls_pivot` по продолжительности и количеству звонков.

#### Звонки

*Примечание:*
«Мегалайн» всегда округляет секунды до минут. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.
* Добавим колонку `month`, которая будет содержать номер месяца.
* Округлим все значения в колонке duration до ровных значений минут и сохраним их в колонке `calls_duration_round`.


```python
# Добавим колонку month, которая будет содержать номер месяца
calls['month'] = pd.DatetimeIndex(calls['call_date']).month
print(calls['month'].unique())
display(calls.head())
```

```python
# Создадим новую колонку с округленными значениями для продолжительности минут
calls['calls_duration_round'] = np.ceil(calls['duration']).astype('int')
calls.head()
```

```python
# Посчитаем суммарное количество минут разговоров и количества звонков за каждый месяц для каждого пользователя
calls_pivot = (calls.
               pivot_table(index=['user_id', 'month'], 
                  values=['id', 'calls_duration_round'],
                  aggfunc={'id':'count', 'calls_duration_round' : 'sum'})
               .reset_index()\
               .rename(columns={'id': 'count_calls_month', 
                                          'calls_duration_round' : 'duration_calls_month'}) #Переменуем названия столбцов

              )
calls_pivot
```

* Проверим на адекватность значения в колонке `duration_calls_month` по суммарному количеству минут по каждому пользователю в месяц.
* Посмотрим, сколько всего было минут разговоров в одном месяце, например, в июне.

```python
# Сколько всего было пользователей в июне
june_count = calls_pivot.query('month==6')['duration_calls_month'].count()
print('Сколько всего было пользователей в июне', june_count)

# Сколько было суммарно минут разговоров в июне
june_call_duration = calls_pivot.query('month==6')['duration_calls_month'].sum()
print('Сколько всего было пользователей в июне', june_call_duration )

# Сколько длился в среднем один разговор в июне
print('Средняя продолжительность одного разговора в июне',round((june_call_duration/june_count)/30), 'минут')
```

*Данные выглядят правдоподобно, 14 минут за разговор это адекватное значение.*

*Вывод*

* Таблица `calls_pivot` содержит суммарное количество минут разговоров и количества звонков за каждый месяц для каждого пользователя.

#### Сообщения

* Создадим сводную таблицу `messages_pivot` по количеству сообщений за месяцам для каждого пользователя.

```python
# Добавим колонку month, которая будет содержать номер месяца
messages['month'] = pd.DatetimeIndex(messages['message_date']).month
print(messages['month'].unique())
display(messages.head())
```

* Создадим сводную таблицу `messages_pivot` по количеству сообщений за месяцам для каждого пользователя.

```python
#Посчитаем суммарное количество сообшений за каждый месяц для каждого пользователя
messages_pivot = (messages.
               pivot_table(index=['user_id', 'month'], 
                  values=['id'],
                  aggfunc={'id':'count'})
               .reset_index()\
               .rename(columns={'id': 'count_messages_month'})   

              )
messages_pivot.head()
```

*Вывод*

#### Интернет

*Примечание:*
«Мегалайн» всегда округляет мегабайты — до гигабайт.
Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.
* Добавим колонку `month`, которая будет содержать номер месяца.
* Создадим сводную таблицу `internet_pivot` по количеству мегабайт трафика.


```python
# Добавим колонку month, которая будет содержать номер месяца
internet['month'] = pd.DatetimeIndex(internet['session_date']).month
print(internet['month'].unique())
display(internet.head())
```

```python
# Посчитаем суммарное количество мгб за каждый месяц для каждого пользователя
internet_pivot = (internet.
               pivot_table(index=['user_id', 'month'], 
                  values=['mb_used'],
                  aggfunc={'mb_used':'sum'})
               .reset_index()\
               .rename(columns={'mb_used': 'mb_used_month'}) # Переименум колонку mb_used
              )
internet_pivot.head()
```

```python
# Округлим значения колонки mb_used до Гб
internet_pivot['gb_used_month'] = np.ceil(internet_pivot['mb_used_month']/1000).astype('int')
internet_pivot = internet_pivot.drop(['mb_used_month'], axis=1) # Удалим лишнюю колонку 'mb_used_month'
internet_pivot.head()
```

```python
# Посмотрим на распределение трафика
print(internet_pivot.gb_used_month.describe())
plt.figure(figsize=(8,6))
sns.boxplot(y=internet_pivot['gb_used_month'])
plt.ylabel('Суммарный объём трафика за месяц, Гб', fontsize=12)
plt.title('Распределение суммарного использованияи интернет-трафика', fontsize=16)
plt.show()
```

* Использование трафика интернета в среднем 17 гб за месяц по обоим тарифам. 
* Минимальное значение 0 мгб - видимо не было использования трафика вовсе.
* Максимальное доходило до 50 Гб.

*Вывод*

* Таблица `internet_pivot` содержит информацию о суммарном количестве Гб за каждый месяц для каждого пользователя.

#### Создание единой таблицы по клиентам

* **Объединим таблицы `calls_pivot`, `messages_pivot`, `internet_pivot` и `users`**


```python
#Посмотрим ещё раз на объемы сводных таблиц
print(calls_pivot.shape)
print(messages_pivot.shape)
print(internet_pivot.shape)
```

## Расчёты и добавление результатов в таблицу

* Рассчитаем помесячную выручку с каждого пользователя.


```python
# Посмотрим на таблицу с тарифами
tariffs
```

```python
#Посмотрим на названия колонок в таблице
users_full.columns
```

```python
# Создадим функцию, которая пройдет по каждой строчке таблицы
# В случае если абонетская плата не превышена лимитов, то таблица верент абонетскую плату
# В случае, если лимит превышен, то к абонетской плате будет прибавлено цена за превышения за звонки, сообшения и интернет
def price(row):
    extra_duration_call = max(row.duration_calls_month - row.minutes_included, 0)*row.rub_per_minute
    extra_messages = max(row.count_messages_month - row.messages_included, 0)*row.rub_per_message
    extra_internet = max(row.gb_used_month - row.gb_per_month_included, 0)*row.rub_per_gb
    return row.rub_monthly_fee + extra_duration_call + extra_messages + extra_internet
```

```python
# Применим функцию price для расчёта стоимости платы по каждому абоненту. Данные будут хранится в новом столбце price
users_full['price'] = users_full.apply(price, axis=1)
users_full.head()
```

* Посмотрим на процентное соотношение переплаты пользователей по тарифам
* Для этого создадим новую колонку `extra_price`, которая будет хранить информацию о переплате в двоичной системе.
* 1 будет означать есть переплата, а 0, что её нет.

```python
# Вызовим ещё раз таблицу по тарифам
tariffs
```

```python
# Проверим координаты
tariffs.loc[1, 'rub_monthly_fee']
```

```python
# Тестирование функции, которая вернет 1, если абонентская плата была превышена и 0, если не превышена

def extra_price(row):
    tariff = row['tariff']
    price = row['price']
    smart_fee = tariffs.loc[0, 'rub_monthly_fee']
    ultra_fee = tariffs.loc[1, 'rub_monthly_fee']
    
    if tariff == 'smart':
        if price>smart_fee:
            return 1
    if tariff == 'smart':
        if price<=smart_fee:
            return 0
        
    if tariff == 'ultra':
        if price>ultra_fee:
            return 1
    if tariff == 'ultra_fee ':
        if price<=ultra_fee:
            return 0
    else:
        return 0
    
      
row_values = ['smart', 551]
row_columns = ['tariff', 'price']
row = pd.Series(data=row_values, index=row_columns) 
print(extra_price(row))  


row_values = ['ultra', 1950]
row_columns = ['tariff', 'price']
row = pd.Series(data=row_values, index=row_columns) 
print(extra_price(row)) 
```

```python
# Создадим колонку 'extra_price', которая будет хранить информацию о наличии переплаты абоненсткой платы
users_full['extra_price'] = users_full.apply(extra_price, axis=1)
```

```python
#Проверим новую колонку 
users_full.tail()
```

* Посмотрим на процентное соотношение по переплате абонетской платы для тарифов, где 0 означает нет переплаты, а 1 есть переплата.

```python
# Посмотрим на переплату для каждого тарифа
def pie_chart(tariff):
    labels = ['Нет переплаты', 'Есть переплата']
    df = users_full[users_full['tariff'] == tariff]
    ax = df.groupby('extra_price').count()\
    .plot(kind='pie',
          y='price',
          autopct='%1.1f%%',          
          figsize=(8, 6),
         labels=labels,
         colors=['#66b3ff', '#ff9999'],
         startangle=90)  
    
    
    # Рисуем donut
    centre_circle = plt.Circle((0,0),0.70,fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)

# Убеждаемся в правильном соотношении сторон  
    ax.axis('equal') 
    plt.tight_layout()
    plt.title('Переплата по тарифу {}'.format(tariff), fontsize=16)
    plt.ylabel('')
    plt.show()
```

## Проверка гипотез

#### Гипотеза №1: Выручка между тарифами
* Сформулируем нулевую гипотезу, что средняя выручка пользователей тарифов «Ультра» и «Смарт» не различается.

* Альтернативная гипотеза, что выручка по тарифу «Смарт» будет меньше, чем по тарифу «Ультра».


```python
#Объемы выборок
print('Объем выборки по тарифу "smart"', len(users_full.query('tariff=="smart"')['price']))
print('Объем выборки по тарифу "ultra"', len(users_full.query('tariff=="ultra"')['price']))
```

* В ходе описательного анализа была обнаружено, что среднее и медианное значение выручки было выше по тарифу `ultra`.
* Проверяем одностороннюю гипотезу, что тариф `smart` **предоставляет меньше выручки, чем тариф `ultra`**.
* Проверка нормальности распределенения не была пройдена, поэтому целесообразние использовать непараметрический метод `mann whitney u test` для неравных независимых выборок с ненормальным распределением данных.

```python
# Проверим гипотезу, что тариф smart приносит меньше выручки тарифа ultra
alpha = .05 # уровень значимости
results = stats.mannwhitneyu(x=users_full.query('tariff=="smart"')['price'],
                   y=users_full.query('tariff=="ultra"')['price'],
                  alternative='less') # результаты непарметрического теста

smart_price_median = users_full.query('tariff=="smart"')['price'].median()
ultra_price_median = users_full.query('tariff=="ultra"')['price'].median()

print('Медианное значение выручки по Смарту', smart_price_median)
print('Медианное значение выручки по Ультра', ultra_price_median)
print(f'p-значение {results.pvalue}')
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу, что выручки между тарифами равны")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Вывод**

* Отвергаем нулевую гипотезу, что выручки между тарифами равны.
* Выручка между тарифами различается и в среднем тариф `smart` приносит меньше дохода, чем тариф `ultra`.

#### Гипотеза №2: средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

* Сформулируем нулевую гипотезу, что средняя выручка пользователей из Москвы и других регионов не различается.

```python
#Объемы выборок по Москве
print('Объем выборки по Москве', len(users_full.query('city=="Москва"')['price']))
print('Объем выборки по другим регионам', len(users_full.query('city!="Москва"')['price']))
```

```python
# Проверка распределений на нормальность при помощи теста Колмогорова-Смирнова для колонок:
# Если распределение нормальное, то уровень значимости (2-е значение в строке) будет больше 0,05.

print(stats.kstest(users_full.query('city=="Москва"')['price'],'norm'))
print(stats.kstest(users_full.query('city!="Москва"')['price'],'norm'))
```

* Медианное значение выручки по Москве выше на 120 рублей. Однако максимальные значения выше по регионам, чем по Москве.
* Объемы выборок очевидно различаются, значит мы будем снова использовать непараметрический тест для разного объема выборок.

```python
# Проверим гипотезу, что тариф smart дешевле тарифа ultra
alpha = .05 # уровень значимости
results = stats.mannwhitneyu(x=users_full.query('city=="Москва"')['price'],
                   y=users_full.query('city!="Москва"')['price'],
                  alternative='two-sided') # результаты непараметрического теста

price_moscow = users_full.query('city=="Москва"')['price'].median()
price_region = users_full.query('city!="Москва"')['price'].median()

print('Медианное значение выручки по Москве', price_moscow)
print('Медианное значение выручки по другим регионам', price_region)
print(f'p-значение {results.pvalue}')
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу, что выручки между Московой и регионами равны")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Вывод**
