# Телеком — отток клиентов

**Цель**

* Научиться прогнозировать отток клиентов телеком компании.

**Задачи**
* Провести предобработку данных.
* Описать поведенческие профили активных и ушедших клиентов.
* Построить классификационную модель для предсказания класса клиента.




**Описание данных**

Собраны исторические данные о договорах, персональной информации, тарифах и услугах.
Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
- `personal.csv` — персональные данные клиента;
- `internet.csv` — информация об интернет-услугах;
- `phone.csv` — информация об услугах телефонии.

*Информация о договорах актуальна на 1 февраля 2020.*


**Описание полей данных**:
* BeginDate – дата начала пользования услугами,
* EndDate – дата окончания пользования услугами,
* Type – тип оплаты: ежемесячный, годовой и тд,
* PaperlessBilling – безналичный расчет,
* PaymentMethod – способ оплаты,
* MonthlyCharges – ежемесячные траты на услуги,
* TotalCharges – всего потрачено денег на услуги
* Dependents – наличие иждивенцев
* Senior Citizen – наличие пенсионного статуса по возрасту
* Partner – наличие супруга(и)
* MultipleLines – наличие возможности ведения параллельных линий во время звонка

**Описание услуг**

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

```python
# Создадим функцию по замене названий колонок
def columns_name(df):
    df.columns = (df\
                    .columns.str.replace( r"([A-Z])", r" \1") # Добавим пропуски между словами
                    .str.strip().str.lower() # Приведем названия колонок к нижнему регистру
                    .str.replace("[( )]", "_")# Добавим _ между словами
                   
                 )
    
    # Переименуем колонку customer_i_d
    df = df.rename(columns={'customer_i_d': 'customer_id'})
    return df
```

```python
# Отфильтруем таблицу без пропусков в колонке total_charges и переведем колонку в формам float
contract = contract.query('total_charges!=" "')
contract['total_charges'] = contract['total_charges'].astype('float')
contract.info()
```

```python
# Извлечём колонку год, чтобы проверить адекватность данных. total_charges должны снижаться с каждым годом.
contract['year'] = pd.DatetimeIndex(contract['begin_date']).year
```

```python
# Построим таблицу для `monthly_charges` и `total_charges` по годам с медианными значениями.
charges = (contract
 .pivot_table(index='year',
                     values=['monthly_charges', 'total_charges'], aggfunc={'monthly_charges': 'median', 
                                                                           'total_charges':'median'}).reset_index()
)


# Посмотрим на распределения цен
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=False)
fig.suptitle('РАСПРЕДЕЛЕНИЯ ЦЕН ПО КОНТРАКТАМ ПО ГОДАМ')

# Месячная оплата
sns.barplot(ax=axes[0], 
            x='year', y='monthly_charges', data=charges, color='#66b3ff')

axes[0].set_title('Месячная оплата')
axes[0].set_ylabel('Медиана')
axes[0].set_xlabel('')


# Годовая оплата
sns.barplot(ax=axes[1], 
            x='year', y='total_charges', data=charges, color='#ff9999')

axes[1].set_title('Итоговая оплата')
axes[1].set_ylabel('Медиана')
axes[1].set_xlabel('')

plt.show()
```

* Медианное значение месячной оплаты имеет нисходящий тренд по годам. Самое высокое значение было в 2013 году.
* Медианное значение итоговой оплаты закономерно каждый год, потому что колонка расчитывается как накопительные данные за весь период пользования контракта.
* Посмотрим на суммарное распределение.

```python
# Построим таблицу для `monthly_charges` и `total_charges` по годам с суммарными значениями.
charges = (contract
 .pivot_table(index='year',
                     values=['monthly_charges', 'total_charges'], 
              aggfunc={'monthly_charges': 'sum',
                       'total_charges':'sum'}).reset_index()
)


# Посмотрим на распределения цен
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=False)
fig.suptitle('РАСПРЕДЕЛЕНИЯ ЦЕН ПО КОНТРАКТАМ ПО ГОДАМ')

# Месячная оплата
sns.barplot(ax=axes[0], 
            x='year', y='monthly_charges', data=charges, color='#66b3ff')

axes[0].set_title('Месячная оплата')
axes[0].set_ylabel('Суммарное значение')
axes[0].set_xlabel('')


# Годовая оплата
sns.barplot(ax=axes[1], 
            x='year', y='total_charges', data=charges, color='#ff9999')

axes[1].set_title('Итоговая оплата')
axes[1].set_ylabel('Суммарное значение')
axes[1].set_xlabel('')

plt.show()
```

* Суммарные значения по месячной оплате повторяют паттерн по клиентам в целом.
* Итоговая оплата закономерно снижается.

### Добавление переменных и объединение таблиц

**Добавление переменных**

```python
# Извлечём месяц из колонки `begin_date`
contract['month'] = pd.DatetimeIndex(contract['begin_date']).month
contract.head()
```

* Создана колонка `month`.
* Даты в колонке `end_date` содержат информацию о времени без уточнений, поэтому проще перевести эти данные в формат даты и посчитать продожительность контракта `contract_duration`.
* Также  создадим целевую колонку `target`, которая будет хранить информацию об оттоке клиентов, где активный договор это 0, а завершившийся контракт - это 1. 
* Представленные данные актуальны на 1 февраля 2020. Заменим все No на эту дату, чтобы потом создать две колонки - ушёл/не ушёл и продолжительность контракта.

```python
# Создадим временную колонку end_date_2, где вместо No будет дата '2020-02-01'
contract['end_date_full'] = contract['end_date'].apply(lambda x: '2020-02-01' if x=="No" else x)
contract['end_date_full'] = pd.to_datetime(contract['end_date_full'], format='%Y-%m-%d')

# Посчитаем продолжительность контракта для каждого пользователя
contract['contract_duration'] = contract['end_date_full'] - contract['begin_date']
contract['contract_duration'] = contract['contract_duration'].dt.days


# Создадим колонку target, которая будет хранить информацию об активных и ушедщих клиентах.
contract['target'] = contract['end_date'].apply(lambda x: 0 if x=="No" else 1)
contract = contract.drop('end_date', axis=1) # удалим лишнюю колонку
contract.head()
```

```python
# Удалим лишнюю колонки begin_date и end_date
contract = contract.drop(['begin_date', 'end_date_full'], axis=1)
contract.head(2)
```

* Cоздана целевая колонка `target`, которая хранить информацию об оттоке клиентов, где активный договор это 0, а завершившийся контракт - это 1. 
* Cоздана колонка `contract_duration`, которая хранить информацию о продолжительности контракта.

**Объединение таблиц**

* Объединим таблицы персональной информации и контрактов в единую таблицу `contract_personal`
* Уберем `paperless_billing`, которая не несёт заведомо важной информации.

```python
contract_personal = (contract
                       .merge(personal, on='customer_id', how='left')
                       .drop(['paperless_billing'], axis=1)
                      )
print(contract_personal.shape)
contract_personal.head()
```

```python
# Объединим таблицы phone и internet
phone_internet = phone.merge(internet, on='customer_id', how='outer')
print(phone_internet.shape)
phone_internet.head(10)
```

* В таблице появились пропуски, т.к. у пользователей разных тарифах не могли быть услуги из другого тарифа.
* Создадим заглушки: для альтернативного тарифа поставим 0. А для использования услуг везде поставим "No".


```python
# Создадим список колонок, где пропуски будут заменены "No"
col = ['multiple_lines', 'internet_service',
       'online_security', 'online_backup', 'device_protection', 'tech_support',
       'streaming_tv', 'streaming_movies']

# Заменим пропуски в выделенных колонках на "No"
phone_internet[col] = phone_internet[col].fillna('No')

# Заменим пропуски в колонках 'phone' и 'internet' на 0
phone_internet = phone_internet.fillna(0)

# Поменяем формат данных колонок 'phone' и 'internet' на 'int'
phone_internet = phone_internet.astype({'phone':"int64",
                                       'internet':"int64"})
phone_internet.shape
```

```python
# Удалим строки с пользователями из списка "customers_na", которые содержат пропуски в колонке 'total_charges'
phone_internet = phone_internet.query('customer_id not in @customers_na')

# Посмотрим информацию в таблице
phone_internet.info()
```

* Все пропуски в объединенной таблице по тарифам заполнены и исправлен формат данных. Удалены строки с клиентами с пропусками в колонке `total_charges`.
* Объединим таблицу `phone_internet` c таблицей `contract_personal` по клиентам.

**Создание единой таблицы**

```python
full_telecom = phone_internet.merge(contract_personal, on='customer_id', how='left')
print(full_telecom.shape)
full_telecom.head()
```

Однако, т.к.. медианные значения для активных клиентов выше, то сформулируем альтернативную гипотезу - cредняя продолжительность контрактов выше у активных клиентов.



```python
# Проверим на нормальность распределения данных о продожительности контрактов с помощью теста Колмогорова-Смирнова
stats.kstest(full_telecom['contract_duration'],'norm')
```

* Данные распределены ненормально, значит будем применять непараметрический тест Манна-Уитни для попарного сравнения двух независимых выборок.

```python
# Проверим одностороннюю гипотезу, что продолжительность контрактов выше у активных пользователей   
alpha = .05 # уровень значимости
results = stats.mannwhitneyu(x=full_telecom.query('target==0')['contract_duration'],
                   y=full_telecom.query('target==1')['contract_duration'],
                  alternative='greater') # результаты непарметрического теста

active_median = full_telecom.query('target==0')['contract_duration'].median()
churn_median = full_telecom.query('target==1')['contract_duration'].median()

print('Медианное значение продолжительности контрактов активных пользователей', active_median)
print('Медианное значение продолжительности контрактов ушедших пользователей', churn_median)
print(f'p-значение {results.pvalue}')
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу, что медианные значения продолжительности контрактов одинаковые")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

* Сформулируем нулевую гипотезу: cредняя месячная и годовая оплаты между активными и ушедшими клиентами одинаковые.
* Однако, т.к. медианные значения для ушедших клиентов выше по месячных расчётах и ниже для итоговых, то сформулируем альтернативную гипотезу - cредняя оплата различается между группами.

```python
# Проверим на нормальность распределения данных об оплатах с помощью теста Колмогорова-Смирнова
col = ['monthly_charges', 'total_charges']
for c in col:
    print(c, stats.kstest(full_telecom[c],'norm'))
```

* Данные распределены ненормально, значит будем применять непараметрический тест Манна-Уитни для попарного сравнения двух независимых выборок.

**Помесячная оплата**

```python
# Проверим одностороннюю гипотезу, что месячная оплата ниже у активных пользователей   
alpha = .05 # уровень значимости
results = stats.mannwhitneyu(x=full_telecom.query('target==0')['monthly_charges'],
                   y=full_telecom.query('target==1')['monthly_charges'],
                  alternative='less') # результаты непарметрического теста

active_median = full_telecom.query('target==0')['monthly_charges'].median()
churn_median = full_telecom.query('target==1')['monthly_charges'].median()

print('Медианное значение месячной оплаты активных пользователей', active_median)
print('Медианное значение месячной оплаты ушедших пользователей', churn_median)
print(f'p-значение {results.pvalue}')
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу, что медианные значения месячной оплаты одинаковые")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

* Активные пользователи достоверно меньше платят в 1,23 раза, чем ушедшие клиенты.

**Итоговая оплата**

```python
# Проверим одностороннюю гипотезу, что итоговая оплата выше у активных пользователей   
alpha = .05 # уровень значимости
results = stats.mannwhitneyu(x=full_telecom.query('target==0')['total_charges'],
                   y=full_telecom.query('target==1')['total_charges'],
                  alternative='greater') # результаты непарметрического теста

active_median = full_telecom.query('target==0')['total_charges'].median()
churn_median = full_telecom.query('target==1')['total_charges'].median()

print('Медианное значение итоговой оплаты активных пользователей', active_median)
print('Медианное значение итоговой оплаты ушедших пользователей', churn_median)
print(f'p-значение {results.pvalue}')
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу, что медианные значения итоговой оплаты одинаковые")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

* Итоговая оплата была достоверно выше у активных пользователей в 2,4 раза, чем у ушедших клиентов.

## Построение моделей

### Подготовка данных перед обучением

* Построим Pipelines для разных моделей классификации. Для этого выделим отдельно количественные и качественные признаки.

```python
# Посмотрим на количественные признаки
full_telecom.select_dtypes(include=['int', 'float']).columns
```

* Среди признаков встречаются и качественные признаки, закодированные цифрами. Отберем только количественные.

```python
numeric_features = ['monthly_charges', 'total_charges', 'contract_duration']
```

```python
# Посмотрим на качественные признаки
full_telecom.select_dtypes(include=['object']).columns
```

* Добавим к признакам 'phone', 'internet'.

```python
categorical_features = ['multiple_lines', 'internet_service', 'online_security',
       'online_backup', 'device_protection', 'tech_support', 'streaming_tv',
       'streaming_movies', 'type', 'payment_method', 'gender', 'partner',
       'dependents', 'phone', 'internet']
```

* Выделим нашу целевую переменную `target` и остальные признаки как features.

```python
target = full_telecom['target']
features = full_telecom.drop(['target', 'customer_id'], axis=1)
```

### Обучение моделей

#### LogisticRegression

```python
# Pipeline 
numeric_transformer = Pipeline(
    steps=[ ("scaler", StandardScaler())]
)

categorical_transformer = OneHotEncoder(handle_unknown="ignore", drop="first")

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

```python
%%time

X_train, X_test, y_train, y_test = train_test_split(features, 
                                                    target,
                                                    test_size=0.25,
                                                    stratify=target, # Разделяем выборку с учётом дисбаланса классов
                                                    random_state=80822)




lr = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', LogisticRegression(random_state=80822,
                                                        solver='liblinear',
                                                        class_weight='balanced'))]) # Борьба с дисбалансом

# Проведем тестирование Логистической регрессии

cross_val_score(lr, X_train, y_train, cv=3, scoring='roc_auc').mean()
```

* Логистическая регрессия демонстрирует высокую оценку `roc-auc` 0,84.

#### CatBoost

* Проведем поиск гиперпараметров через CatBoost.
* Этот алгоритм не требует предварительного преобразования данных, а лишь указать качественные переменные.

```python
# Создадим новую матрицу признаков и переведем формат object в category.
features_boost = features
features_boost[categorical_features] = features_boost[categorical_features].astype('category')
```

```python
features_boost.info()
```

* Данные преобразованы.
* Разобьем наши признаки на тренировочную и тестовую выборки.

```python
X_train, X_test, y_train, y_test = train_test_split(features_boost, target, 
                                                    stratify=target,
                                                    test_size=0.25, 
                                                    random_state=80822)
```

```python
# Проведем поиск гиперпараметров через RandomizedSearchCV c scoring='roc_auc'

model_cb = CatBoostClassifier(auto_class_weights = 'Balanced')

# Гиперпараметры
param_dist = {'learning_rate': [0.03, 0.1],
        'depth': [4, 6, 10],
          'iterations':[500, 1000, 1500],
          'metric_period':[10,50],
            'od_wait': [10,30],
            'od_type':['None', 'Iter']}
               
# RandomizedSearchCV
cb_RandomGrid = RandomizedSearchCV(model_cb,
                                   param_dist, 
                                   scoring='roc_auc', 
                                   cv =3, 
                                   verbose=100,
                                   random_state=80822)
```

```python
%%time
# Запускаем поиск гиперпараметров
cb_RandomGrid.fit(X_train, y_train, cat_features=categorical_features)
```

```python
print(cb_RandomGrid.best_params_)
print(cb_RandomGrid.best_score_)
```

* Модель демонстрирует высокую оценку 89,9%.

## Тестирование лучшей модели

```python
%%time
# Обучим модель с лучшими гиперапараметрами

model_cb = CatBoostClassifier(**cb_RandomGrid.best_params_)

# Создадим пулы данных для выборок
train_pool = Pool(X_train, y_train, cat_features=categorical_features)



model_cb.fit(
    train_pool,
    plot=True,
    use_best_model=True,

)
```

```python
# Получим предсказание
target_names = ['Клиент не ушёл', 'Клиент ушёл']
prediction_cb = model_cb.predict(X_test)
print(classification_report(y_test, prediction_cb, target_names=target_names))
```
